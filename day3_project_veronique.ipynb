### Split 'trans_date_trans_time' into date and time

```python
# Split 'trans_date_trans_time' into date and time (convert the data type)

cc_df['trans_date'] = cc_df['trans_date_trans_time'].str.split().apply(lambda x: x[0])
cc_df['trans_date'] = pd.to_datetime(cc_df['trans_date'], format='%Y-%m-%d')

cc_df['trans_time'] = cc_df['trans_date_trans_time'].str.split().apply(lambda x: x[-1])
cc_df['trans_time'] = pd.to_timedelta(cc_df['trans_time'])
```

### Change the data type for 'dob' 

```python
# Change the data type for 'dob' 

cc_df['dob'] = pd.to_datetime(cc_df['dob'], format='%Y-%m-%d')
```

### Delete 'fraud_' string from 'merchant'

```python
# Replace 'fraud_' from 'merchant'

cc_df['merchant'] = cc_df['merchant'].str.replace('fraud_','')
```

### Rename columns to indicate customer-specific information

```python
# Return the names of columns in Series

pd.Series(cc_df.columns)
```

```python
# Generate a list of columns to rename

rename_col = list(cc_df.columns[5:14]) + list(cc_df.columns[[15,16]])
```

```python
# Confirm if the list is complete

rename_col
```

```python
# Function to rename the columns

new_col = []

for col in cc_df.columns: 
    if col in rename_col: 
        col = 'c_' + col
    else:
        col
    new_col.append(col)
```

```python
# Check the new column names

new_col
```

```python
# Rename the columns

cc_df.columns = new_col
```

### Age

The customer's age as of the latest transaction date

```python
# Find the latest transaction

latest_tran = pd.to_datetime(ft_df['trans_date'].max())
print('The latest transaction date is', latest_tran.date())
```

```python
# Calculate the age as of the latest transaction date

def age_func(dob):
    dob = pd.to_datetime(dob).date()
    return latest_tran.year - dob.year - ((latest_tran.month, latest_tran.day) < (dob.month, dob.day))   
```

```python
# Create a 'c_age' column containing the age of the customer as of the latest transaction date in the dataset

ft_df['c_age'] = ft_df['c_dob'].apply(age_func)
```

```python
# Drop the 'c_dob' as it's now obsolete

ft_df.drop('c_dob', axis=1, inplace=True)
```

```python
# Check for any outliers in the new column

ft_df.describe()
```

### Fradulent merchant

Flag merchants with high fraud rate

```python
# Cross-tabulation between 'merchant' and 'is_fraud'

fraud_merch = pd.crosstab(ft_df['merchant'], ft_df['is_fraud']).rename({0:'safe', 1:'fraud'}, axis=1)
```

```python
# Add 'total'

fraud_merch['total'] = fraud_merch['safe'] + fraud_merch['fraud']
```

```python
# Calculate the fraud ratio

fraud_merch['pct'] = fraud_merch['fraud'] / fraud_merch['total'] * 100
```

```python
# Show the top 15 merchants whose business was associated with fraud

fraud_merch.sort_values(by='pct', ascending=False).head(15)
```

```python
# Show the names of merchants with #fraud > 3 and %fraud > 1%

fraud_merch[(fraud_merch['pct'] > 1) & (fraud_merch['fraud'] > 3)]
```

```python
# Save their indices

flag_merch = fraud_merch[(fraud_merch['pct'] > 1) & (fraud_merch['fraud'] > 3)].index
```

```python
# Dummy column marking merchants in flag_merch 

ft_df['flag_merch'] = ft_df['merchant'].isin(flag_merch).astype(int)
```

```python
# Check the first 15 rows flagged 

ft_df[ft_df['flag_merch'] == 1].head(15)
```

```python
# Drop the 'merchant' column

ft_df.drop('merchant', axis=1, inplace=True)
```

### Day of the week

```python
# Import library

import datetime
```

```python
# Keep the weekday inputs in numerical value

ft_df['weekday'] = ft_df['trans_date'].apply(lambda x: x.date().weekday())
```

```python
# Map out weekday numbers to weekday names

weekdays = {0: 'Mon',
            1: 'Tue',
            2: 'Wed',
            3: 'Thu',
            4: 'Fri',
            5: 'Sat',
            6: 'Sun'}
```

```python
# Introduce an additional column representing the weekdays in words (i.e. 'Mon', 'Tue', etc.)

ft_df['weekday_wrd'] = ft_df['weekday'].map(weekdays)
```

```python
# Define the order of the days of the week

weekday_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
```

```python
# Convert the weekday column to a categorical type with the specified order

ft_df['weekday_wrd'] = pd.Categorical(ft_df['weekday_wrd'], categories=weekday_order, ordered=True)
```

```python
# Formulate a DataFrame illustrating the trend in fraud across the days of the week

by_wkday = pd.DataFrame(ft_df.groupby('weekday_wrd')['is_fraud'].sum())
```

```python
# Insert a 'ratio' column

by_wkday['ratio'] = by_wkday.apply(lambda x: x / ft_df['is_fraud'].sum() * 100)
```

```python
# Print by_wkday

by_wkday
```

```python
# Demonstrate with a bar plot

sns.barplot(x=by_wkday.index, y=by_wkday['ratio'])
```

```python
# Rid of 'weekday_wrd'

ft_df.drop('weekday_wrd', axis=1, inplace=True)
```

### Latitude/longitude

```python
# Insert new columns investigating the distance between the customer's address and that of the merchant in terms of lattitude/longitude

ft_df['long_diff'] = abs(ft_df['c_long'] - ft_df['merch_long'])
ft_df['lat_diff'] = abs(ft_df['c_lat'] - ft_df['merch_lat'])
```

```python
# Drop the old columns

ft_df.drop(['c_lat', 'c_long', 'merch_lat', 'merch_long'], axis=1, inplace=True)
```
